```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movies_metadata.csv')
df.head(5)
```

```python
drop_df = ["adult","belongs_to_collection","budget","genres","imdb_id","original_language","overview","homepage", "poster_path", "video", "imdb_id", "overview", "original_title", 
           "spoken_languages", "tagline","overview","original_title","popularity","production_companies","production_countries","release_date","revenue","runtime","status","video"]

df = df.drop(drop_df, axis=1)
df = df.drop_duplicates(keep='first')
df.dropna(how='all', inplace=True)
df
```

```python
df = df.sort_values(by=['vote_average'], ascending=False)
df
```

```python
vote_counts = df[df['vote_count'].notnull()]['vote_count'].astype('int')
vote_averages = df[df['vote_average'].notnull()]['vote_average'].astype('int')
C = vote_averages.mean()
m = vote_counts.quantile(0.9)
def weighted_rating(data):
    v = data['vote_count']
    R = data['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)

df['weighted_rating'] = df.apply(weighted_rating, axis=1)
```
