```python
# converts json list to list of inputs (from the label specified with 'wanted' parameter)
def json_to_arr(cell, wanted = "name"): 
    cell = literal_eval(cell)
    if cell == [] or (isinstance(cell, float) and cell.isna()):
        return np.nan
    result = []
    counter = 0
    for element in cell:
        if counter < 3:
            result.append(element[wanted])
            counter += 1
        else:
            break
    return result[:3]
```

```python
df[['genres']] = df[['genres']].applymap(json_to_arr)
df[['production_companies']] = df[['production_companies']].applymap(json_to_arr)   
df[['production_countries']] = df[['production_countries']].applymap(json_to_arr)
```

```python
def list_counter(col, limiter = 9999):
    result = dict()
    for cell in col:
        if isinstance(cell, float):
            continue
        for element in cell:
            if element in result:
                result[element] += 1
            else:
                result[element] = 1
    result = {k: v for k, v in sorted(result.items(), key=lambda item: item[1], reverse=True)}
    counter = 1

    returned = []
    for i in result:
        if counter > limiter:
            break 
        counter += 1
        returned.append([i, result[i]])
    return returned
```

```python
genres_count = list_counter(df['genres'].values)
genres = pd.DataFrame.from_records(genres_count, columns=['genres', 'count'])
genres.plot(kind='bar', x='genres')
plt.xlabel('genres')
plt.ylabel('count')
plt.show()
```

```python
companies_count = list_counter(df['production_companies'].values)
companies = pd.DataFrame.from_records(genres_count, columns=['production_companies', 'count'])
companies.plot(kind='bar', x='production_companies')
plt.xlabel('production_companies')
plt.ylabel('count')
plt.show()
```

```python
countries_count = list_counter(df['production_countries'].values, 20)
countries = pd.DataFrame.from_records(countries_count, columns=['production_countries', 'count'])
countries.plot(kind='bar', x='production_countries')
plt.xlabel('production_countries')
plt.ylabel('count')
plt.show()
```

```python
def fill_na_with_list(cell, data):
    if isinstance(cell, float):
        return data
    else:
        return cell
```

```python
df[['genres']] = df[['genres']].applymap(lambda row:
                                        fill_na_with_list(row, [genres['genres'][0]]))
df[['production_countries']] = df[['production_countries']].applymap(lambda row: 
                                        fill_na_with_list(row, [countries['production_countries'][0]]))
df[['production_companies']] = df[['production_companies']].applymap(lambda row: 
                                        fill_na_with_list(row, [companies['production_companies'][0]]))
```

```python
df['profit'] = df['revenue'] - df['budget']
```

```python
min_val = df["budget"].min()
max_val = df["budget"].max()
df[["budget", "revenue", "profit"]] = df[["budget", "revenue", "profit"]].apply(lambda x: 
                                                            x / (max_val - min_val))
```

```python
vote_counts = df[df['vote_count'].notnull()]['vote_count'].astype('int')
vote_averages = df[df['vote_average'].notnull()]['vote_average'].astype('int')
C = vote_averages.mean()
m = vote_counts.quantile(0.75)
def weighted_rating(data):
    v = data['vote_count'] + 1 # added +1
    R = data['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)

df['weighted_rating'] = df.apply(weighted_rating, axis=1)
```

```python
def get_director(x):
    x = literal_eval(x)
    for i in x:
        if i == "[]" or isinstance(i, float):
            return np.nan
        if i['job'] == 'Director':
            return i['name']
    return np.nan

df['director'] = df['crew'].apply(get_director)
df.drop(["crew"], axis=1, inplace=True)
```
