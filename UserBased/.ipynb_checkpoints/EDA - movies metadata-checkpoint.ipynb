```python
def get_list(x, l=3):
    if isinstance(x, list):
        if len(x) > l:
            x = x[:l]
        return x

    return []
```

```python
df['genres'] = df['genres'].apply(ast.literal_eval)
df['production_companies'] = df['production_companies'].apply(ast.literal_eval)
df['production_countries'] = df['production_countries'].apply(ast.literal_eval)
df['cast'] = df['cast'].apply(ast.literal_eval)
df['keywords'] = df['keywords'].apply(ast.literal_eval)
```

```python
df['genres'] = df['genres'].apply(get_list)
df['production_companies'] = df['production_companies'].apply(get_list)
df['production_countries'] = df['production_countries'].apply(get_list)
df['cast'] = df['cast'].apply(get_list)
df['keywords'] = df['keywords'].apply(get_list)
```

```python
def build_plots_list(data, target=None):
    tmp = data[['title', target]].astype(str).drop_duplicates()
    tmp[target] = tmp[target].apply(ast.literal_eval)
    count_dict = {}
    for t in tmp[target]:
        for g in t:
            count_dict[g] = count_dict.get(g, 0) + 1
    count_data = pd.DataFrame.from_dict(data={target: count_dict.keys(), 'count': count_dict.values()})
    count_data = count_data.sort_values(ascending=False, by='count')
    sns.barplot(x= count_data[target][:10],
           y= count_data['count'][:10])

    plt.xlabel(target)
    plt.ylabel('Count')
    plt.title('Show of {} Bar Plot'.format(target))
    plt.xticks(rotation=90)
    plt.show()
```

```python
pd.DataFrame((df.sort_values('vote_count',ascending=False)['title']).unique())[:10]
```

```python
build_plots_list(data=df, target='genres')
build_plots_list(data=df, target='production_companies')
build_plots_list(data=df, target='production_countries')
```
